```python
# (POR) Verificando nulos nas colunas
# (ENG) Checking nulls in each column

df.isnull().sum()
```

```python
# (POR) Plotando gráfico de calor ( cada linha representa um valor nulo na coluna)
# (ENG) Plotting heat graph (each line represents a null value in the column)

plt.figure( figsize = (15,6) )
plt.title('Analisando Campos Nulos')
sns.heatmap( df.isnull() )
```

```python
# (POR) Contando valores unicos em cada coluna
# (ENG) Counting unique values ​​in each column

df.nunique()
```

```python
# (POR) Checando quais os valores unicos na coluna "Industry" para saber quais os ramos das empresas
# (ENG) Checking the unique values ​​in the "Industry" column to find out which branches of companies

df['Industry'].unique()
```

```python
# (POR) Contando quantas empresas por tipo de industria
# (ENG) Counting how many companies per type of industry
df['Industry'].value_counts()
```

```python
# (POR) Criando a variavl top10 tendo em vista o quanto o ramo representa no total
# (ENG) Creating the top10 variable taking into account how much the branch represents in total

top_10 = round(df['Industry'].value_counts( normalize = True) *100 , 2).head(10)

display(top_10)
```

```python
# (POR) Plotando gráfico mostrando cada industria
# (ENG) Plotting a graph showing each industry

sns.set_style("whitegrid")
plt.figure ( figsize= (15,6))
plt.title ( ' Analise de Setores')
plt.bar( df['Industry'].value_counts().index, df['Industry'].value_counts() )
plt.xticks ( rotation = 45 , ha= 'right' );
```

```python
# (POR) Top10 de representativiade de cada pais
# (ENG) Top10 representativeness of each country

top10_country = round( df['Country'].value_counts( normalize=True ) * 100, 1 ).head(10)
display(top10_country)
```

```python
#parametros
labels = top10_country.index[0:10]
vals = top10_country
explode = (0.05, 0)
colors = '#39B8B8' , '#B8397C'

#plotação do grafico
fig, gr_gender = plt.subplots( figsize= (15,8) )
gr_gender.pie(vals, labels= labels , autopct= '%.1f', shadow = True )
gr_gender.set_title('Top 10 Paises com mais empresas', fontsize = 14)
plt.show
```

```python
# (POR) Conversão para Data
# (ENG) Date conversion
df['Date Joined'] = pd.to_datetime( df['Date Joined'] )

df['Date Joined'].head()
```

```python
# (POR) Formatando e criando as colunas mes e ano para analises
# (ENG) Formatting and creating month and year columns for analysis
df['month'] = pd.DatetimeIndex( df['Date Joined'] ).month
df['year'] = pd.DatetimeIndex( df['Date Joined'] ).year

df.head()
```

```python
# (POR)Tabela Analitica
# (ENG) Analytical Table

data_group = df.groupby( by=['Country', 'year', 'month', 'Company'] ).count()['Id'].reset_index()

data_group

```

```python
# (POR) Usando .loc para filtrar por pais (analisando o Brasil)
# (POR) Using .loc to filter by country (analyzing Brazil)
data_group.loc[
  data_group['Country'] == 'Brazil'
]
```

```python
# (POR) Renomeando a coluna Valuation
# (ENG) Renaming the Valuation column
df['Valuation ($B)'] = pd.to_numeric( df['Valuation ($B)'].apply( lambda Linha: Linha.replace('$', '') ) )

df.head()
```

```python
# (POR) Agrupando para saber o a soma de valuation de cada pais
# (ENG) Grouping to know the sum of valuation of each country
df_country = df.groupby( by=['Country'] ).sum()['Valuation ($B)'].reset_index()

vw_country = df_country.sort_values(by= 'Valuation ($B)', axis=0, ascending=False).head(10)

df_country.head(10)
```

```python
# (POR) Top10 dos paises com maiores valuations
# (ENG) Top 10 countries with the highest valuations
df_price = df_country.sort_values('Valuation ($B)', ascending=False)
df_price.head(10).reset_index()
```
